```python
import torch
import cv2
import os
import pandas as pd
import numpy as np
```

```python
def save_file_names():
    folder_path = os.path.abspath('/Users/cerenmorey/Desktop/BeCode/becode_projects/B&D/becode-signature-object-detection/test')
    all_files = os.listdir(folder_path)
    file_id = []
    for file in all_files:
        test_file_name = file[:-4]
        file_id.append(test_file_name)
    return file_id
```

```python
file_id = save_file_names()
len(file_id)
```

```python
test_images = []
for i, name in enumerate(file_id):

    folder_path = os.path.abspath('/Users/cerenmorey/Desktop/BeCode/becode_projects/B&D/becode-signature-object-detection/test')
    file_path = os.path.join(folder_path, name+'.tif')
    img = cv2.imread(file_path, 0)
    test_images.append(img)
```

```python
len(test_images)
```

```python
model = torch.hub.load('ultralytics/yolov5', 'custom',  path='/Users/cerenmorey/Downloads/best (1).pt')
```

```python
all_data = pd.DataFrame()
```

```python
for i in range(len(test_images)):

    results = model(test_images[i])

    if len(results.pandas().xyxy[0]) > 0:
        temp_all_data = pd.DataFrame()
        temp_all_data = results.pandas().xyxy[0]
        temp_all_data['Id'] = file_id[i]
    #     temp_all_data['Id'] = file_id[i]
        all_data = all_data.append(temp_all_data, ignore_index=True)
    else:
        empty_data = pd.DataFrame([[0,0,0,0,0,100]],columns = ['xmin', 'ymin', 'xmax', 'ymax', 'confidence', 'class'])
        empty_data['Id'] = file_id[i]
        all_data = all_data.append(empty_data, ignore_index=True)
```
